```python
from langgraph.graph import StateGraph, START, END
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv
from typing import TypedDict
from pydantic import BaseModel, Field

load_dotenv()
```

```python
llm = HuggingFaceEndpoint(
    repo_id="Qwen/Qwen3.5-9B",
    task="text-generation"
)

model = ChatHuggingFace(llm=llm)
```

```python
class EvaluationSchema(BaseModel):
    feedback: str = Field(description="The feedback for the essay provided by the model.")
    score: int = Field(description="The score for the essay provided by the model, on a scale of 0 to 10.")
    
```

```python
model.with_structured_output(EvaluationSchema)
```
